```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Step1: Drop the NaN data in 'CPI'
print('Step1: Drop the NaN data in CPI:')
dataset_RetailFeature.dropna(subset=['CPI'])
print(dataset_RetailFeature.isnull().sum())

#Step2: Keep only the rows with at least 5 non-na values::
print('Step2: Keep only the rows with at least 5 non-na values:')
dataset_RetailFeature.dropna(thresh=5)
print(dataset_RetailFeature)

#Step3: Identify placeholder data in 'MarkDown1' (e.g. 0's for NA's):
print('Step3.1: Identify placeholder data in MarkDown1:')
dataset_RetailFeature['MarkDown1'].fillna('0',inplace=True)

print('Step3.2: Fill the NaN data in MarkDown2 -MarkDown5 with mean:')
dataset_RetailFeature['MarkDown2'].fillna(dataset_RetailFeature['MarkDown2'].mean(),inplace=True)
dataset_RetailFeature['MarkDown3'].fillna(dataset_RetailFeature['MarkDown3'].mean(),inplace=True)
dataset_RetailFeature['MarkDown4'].fillna(dataset_RetailFeature['MarkDown4'].mean(),inplace=True)
dataset_RetailFeature['MarkDown5'].fillna(dataset_RetailFeature['MarkDown5'].mean(),inplace=True)
print('Step3: Fill the NaN data in MarkDownX with mean:')
print(dataset_RetailFeature)

#Step4: Identify outliers: 
# Supposed keep only the that are within +3 to -3 standard deviations in the column 'Temperature':
print('Step4: Identify outliers:')
dataset_RetailFeature[np.abs(dataset_RetailFeature['Temperature']-dataset_RetailFeature['Temperature'].mean())
                      <=3*dataset_RetailFeature['Temperature'].std()]
print(dataset_RetailFeature)

#Step5: Check for overall plausibility and errors (e.g., typos, unreasonable ranges)
print('Step5.1: Check for overall plausibility and errors-unreasonable ranges')
dataset_RetailFeature[np.abs(dataset_RetailFeature['Fuel_Price']-dataset_RetailFeature['Fuel_Price'].mean())
                      <=4*dataset_RetailFeature['Fuel_Price'].std()]
print(dataset_RetailFeature)

print('Step5.2: Check for overall plausibility and errors-Typos')
L=pd.Series(['TRUE','FALSE']) #values_not_in_array = df[~df.Suburb.isin(L)].Suburb
dataset_RetailFeature = dataset_RetailFeature[~dataset_RetailFeature.IsHoliday.isin(L)]
print(dataset_RetailFeature)

#Step6: Identify highly correlated variables
print('Step6: Identify highly correlated variables:')
print(dataset_RetailFeature.corr().abs().unstack().sort_values(kind="quicksort"))

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
import seaborn as sns

dataset_RetailFeature=pd.read_csv(r"C:\Users\Zi Wei Fan\Desktop\NEU\NEU\Course\INFO7390_AdavancedDataScience\7390_Assignment\Assignment1\retail-data-analytics\Features_DataSet.csv")
print(dataset_RetailFeature.head(10)) #print head 10 items.

dataset_RetailFeature=dataset_RetailFeature.fillna(dataset_RetailFeature['CPI'].mean())
dataset_RetailFeature=dataset_RetailFeature.fillna(dataset_RetailFeature['Unemployment'].mean())

data=dataset_RetailFeature[["Temperature","Fuel_Price"]]
#data=data.as_matrix().astype("float32",copy=False)

#Detect outlier with DBSCAN:
dbscan=DBSCAN(eps=.5,min_samples=15).fit(data)
labels=dbscan.labels_
coreSamples=np.zeros_like(labels,dtype=bool)
coreSamples[dbscan.core_sample_indices_]=True
labels=dbscan.labels_

#Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)

#Plot:
sns.set(style="ticks", color_codes=True)
unique_labels = set(labels)
colors = [plt.cm.Spectral(each)
          for each in np.linspace(0, 1, len(unique_labels))]

print('Scatter Plot:')
ax1=data.plot.scatter(x='Temperature',y='Fuel_Price',color='DarkBlue',label='Temperature-Fuel_Price')
plt.show()

print('Pairplot:')
#sns.set(style="ticks", color_codes=True)
ax2=sns.pairplot(data)
plt.show()
```

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
from openpyxl import Workbook
import urllib
import requests
import re

#save external links:
html=urlopen("http://www.mfa.org/programs/lectures-and-courses/lectures").read()
soup=BeautifulSoup(html,"html.parser")
#Create xls:
wb = Workbook() # open new workbook, use load_workbook if existing
ws = wb.create_sheet('Sheet_Assignment1Result')

divs=soup.find_all('div',{'class':'grid-4'})
for div in divs:
    dummyrow=[]
    h4s=div.find_all('h4')
    for h4 in h4s:
        dummyrow.append(h4.get_text().strip())
        print(h4.get_text().strip())
        #Save to excel:
        
    aList=div.find_all("a")
    for a in aList:
        link='http://www.mfa.org'+a['href']
        dummyrow.append(link)
        print(link)
        
        r=requests.head('http://www.mfa.org'+a['href'])
        dummyrow.append(r.status_code)
        print(r.status_code)
        ws.append(dummyrow)
wb.save(r'C:\Users\Zi Wei Fan\Desktop\NEU\NEU\Course\INFO7390_AdavancedDataScience\7390_Assignment\Assignment1\Assignment1Result.xlsx')
```
