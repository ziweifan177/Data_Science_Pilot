```python
#importing the libraries for stroing the file in fasta format
#this will make it easy to handle while doing the sequence alignment using clustalw before clustering. 
import os
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio.Alphabet import generic_dna
```

```python
#creating the dictionary "my_records" to store all the 34 sequences. 
my_records = [SeqRecord(Seq(sequence, generic_dna),
              id='Species_{}'.format(letter), description='Species_{}'.format(letter))
              for sequence, letter in zip(sequences, 'abcdefghijklmnopqrstuvwxyz12345678')]
```

```python
#providing the file path to store the fasta file
root_dir = r"\\\\"
filename = 'my_sequences'
fasta_path = os.path.join(root_dir, '{}.fasta'.format(filename))
```

```python
#Writing the sequences to the fasta file.
#still figuring out a way to make it work
SeqIO.write(my_records, fasta_path, "fasta")
```
