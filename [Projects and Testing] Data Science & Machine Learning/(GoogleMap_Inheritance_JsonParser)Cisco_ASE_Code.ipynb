### 1.Write a function that will convert a JSON-formatted string to a dictionary.  
- For instance,the function will take this string as a parameter:  
"{'totalCount':'1','ID':'1029','IP':'10.0.0.1'}"  
And will return a dictionary called “result” that will have the following values:  
result[“totalCount”] = “1” result[“ID”] = “1029” result[“IP”] = “10.0.0.1”

####  <font color = 'red'> Parse by Slice, Pattern matching </font>

<img src='Image/vali_3.png' style="width: 800px;">

```python
import string

def validate_if_validJson(str_json):
    '''This function is used for validating if the json is in valid json format:'''
    if not str_json:                                # Case 1: Input validation
        return False                                #1.1: If no any input: must not a valid input.
    
    elif str_json[0]!='{' or str_json[-1]!='}':     #1.2: if the curly brackets cannot match the beginning and the tail. Not a valid json: "{'a':'1','b':'2']" 
        return False
    
    else:                                           # Case 2: sub pairs validation
        list_subString = str_json[1:-1].split(',')
        print('All pairs in string: ', list_subString)
        
        for pair in list_subString: # Pair checking:
            if pair[0]!='\'' or pair[-1]!='\'':     # 2.1 Some pair didn't start with single quotation mark.
                print("It is invalid Json, because:", pair, " pair[0] is not \' or pair[-1] is not \'. ")
                return False
            
            elif '\':\'' not in pair:               # 2.2 Some pair didn't follow the pattern: ':'
                print("It's invalid Json, because '\':\'' not in pair. ")
                return False
            
            elif '\':\'' in pair:                   # 2.3 Single quotation marks didn't match in some pair.
                count_single_quote = 0
                for char in pair:
                    if char == '\'':
                        count_single_quote +=1
                if count_single_quote%2!=0:         
                    print('It\'s invalid Json, because ', pair, 'single quotation marks do not match.')
                    return False
                
                                                    
            split_key_value = pair.split(':')            # Case 3: Key-value pair Validation:
            print('split_key_value: ', split_key_value)
                
            for item in split_key_value:            
                                                    
                item = item[1:-1]                        # (We just focus on the content between 2 single quotations: 'count':'2', we care about count & 2.)
                print('item: ', item)
                # 3.1 Some pair has invalid value or key.
                if len(item) == 0:                      # 3.1 Some pair has not value or key.
                    print('Not valid, pair: ', pair, 'length is 0.')
                    return False
                
                elif len([s for s in item if s not in string.punctuation])==0:   # 3.2 Some pair only has punctuation in key/value.
                    print('Not valid, pair: ', pair, 'length of valid content 0.')
                    return False
    return True
    
```

```python
str_json = "{'totalCount':'1','ID':'1029','IP':'10.0.0.1'}"
validate_if_validJson(str_json)
```

```python
def generate_Dictionary(str_json):
    if not validate_if_validJson(str_json):
        print('Please validate the json input.')
        return -1
    
    else:
        dict_result = dict()
        
        for pair in str_json[1:-1].split(','):               # Step 2: Pre-process by slice and split
            print('\npair:', pair)
            
            item_split  = pair.split(':')
           
            dict_result[item_split[0]] = item_split[1]      # Step 3: Assign the key and value
            
    return dict_result
```

### 3. Create class definitions that will properly represent the following constructs:  
Define a Shape object, where the object is any two-dimensional figure, and has the following characteristics:  
name (string)  
perimeter (integer)  
area (integer)  

Define a Circle object. A Circle has the characteristics of a Shape object. A Circle is a Shape.  
Define a Triangle object. A Triangle has the characteristics of a Shape object. A Triangle is a Shape. However, a Triangle has the additional characteristic:  
sides = [length1, length2, length3] Note that “sides” is an array of integers.  
Instantiate an object of each type (Shape, Circle, Triangle).

#### <font color = 'red'>Inheritance + Polymorphism</font>

<img src='Image/int.png' style="width: 700px;">

```python
import math
```

#### Let's define the parent class:

```python
class Shape:
    def __init__(self, name='Default_shapeName', perimeter=0, area=0):
        self.name = name
        self.perimeter = perimeter
        self.area = area
        
    def get_name(self):
        return self.name
    
    def get_perimeter(self):
        return self.perimeter
    
    def get_area(self):
        return self.area

    def print_(self):
        print("This is a shape. My name is: ", self.name)
```

#### Define the child class-Circle:  
- Child classes have the same name as the methods in the parent class...

```python
class Circle(Shape):   # derived subclass Circle by inheritance
    def __init__(self, name, perimeter):
        Shape.__init__(self, name, perimeter)
        self.area = (self.perimeter/(2*math.pi))**2*math.pi # area = (perimeter/(2*pi))**2*pi
        
    def print_(self):    # Overwrite.
        print("This is a Circle. My name is: ", self.name)
```

#### Define the child class-Triangle:

```python
class Triangle(Shape):   # derived subclass Triangle by inheritance: Shape(self, name, perimeter, area)  
    def __init__(self, name, sides):
        Shape.__init__(self, name)
        self.sides = sides
        self.perimeter = sum(self.sides) 

    def get_perimeter(self):  # Overwrite here.
        return self.perimeter
    
    def get_area(self):       # Overwrite here + Validate
        if not self.if_Triangle():
            return -1
        
        p = sum(self.sides)/2  
        return math.sqrt(p*(p-self.sides[0])*(p-self.sides[1])*(p-self.sides[2]))  # Caculate area based on 3 sides.       

    def print_(self):        # Overwrite here.
        print("This is a Triangle. My name is: ", self.name)
    
    def if_Triangle(self):   # Validate
        if len(self.sides)==3:
            return True
        return False
```

#### Let's initialize the 3 classes.

```python
# Initialize Shape:
shape = Shape('shape_Tom', 13, 8)               # Initialize a shape.
circle = Circle('circle_Rick', 3)              # Initialize a circle: name, perimeter
triangle = Triangle('triangle_Morty', [3,4,5]) # Initialize a triangle: name, sides
```
