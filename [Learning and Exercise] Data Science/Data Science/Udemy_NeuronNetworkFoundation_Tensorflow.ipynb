#### Step3: Create the model:

- y = W*X+b

- 1. Placeholder:

```python
X = tf.placeholder(tf.float32, shape=[None, 784])
# None: indicates that the first dimension, 
# corresponding to the batch size, can be of any size.
```

```python
type(X)
```

```python
# Matrix Weight:
W = tf.Variable(tf.zeros([784, 10]))
```

```python
# Bias:
b = tf.Variable(tf.zeros([10]))
```

```python
# Create the graph:
y = tf.matmul(X,W)+b
```

```python
y_true = tf.placeholder(tf.float32, [None,10])
#y_true is the real ground truth for matching the y predicted.
# Still could not decide the rows in the batches, so None.
```

- 2. Loss and Optimizer:

    - Multi-Class Neural Networks: Softmax
            Softmax might produce the following likelihoods of an image belonging to a particular class. Softmax is implemented through a neural network layer just before the output layer.
            The Softmax layer must have the same number of nodes as the output layer.  
            Softmax assumes that each example is a member of exactly one class.

- Softmax turn a vector of numbers, called logits, into a probabilities that sum to one: 0.7+0.2+0.1=1

```python
Image(filename='Image/softmax.png')
```

```python
# Cross Entropy:
cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y_true, logits=y))
# numpy.mean and tensorflow.reduce_mean are the same.
```

- Cross-entropy loss, or log loss, measures the performance of a classification model whose output is a probability value between 0 and 1.  
- A perfect model would have a log loss of 0.
- As the predicted probability approaches 1, log loss slowly decreases. As the predicted probability decreases, however, the log loss increases rapidly

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.5)
```

- Gradient Descent: Repeating this process enough times will lead to the bottom of the bowl and you will know the values of the coefficients that result in the minimum cost:
        1. A random position on the surface of the bowl is the cost of the current values of the coefficients (cost).

        2. The bottom of the bowl is the cost of the best set of coefficients, the minimum of the function.

        3. The goal is to continue to try different values for the coefficients, evaluate their cost and select new coefficients that have a slightly better (lower) cost.  
- Cost Function: 
    1. Cost functions are used to estimate how badly models are performing. 
    2. expressed as: a difference or distance between the predicted value and the actual value. 
    3. you may also see 'Cost function' referred to as loss or error.

```python
train = optimizer.minimize(cross_entropy)
```

#### Step4: Create Session:

```python
Image(filename='Image/pro.png')
```

```python
init = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init)
    
    # Train the model for the 1000 steps on the training set;
    # Using built in batch feeder from mnist, for convenience.
    
    for step in range(1000):
        batch_x, batch_y = mnist.train.next_batch(100)
        sess.run(train, feed_dict={X: batch_x, y_true: batch_y})
        
    # Test the train model:
    matches = tf.equal(tf.argmax(y,1), tf.argmax(y_true,1))
    acc = tf.reduce_mean(tf.cast(matches, tf.float32))
    
    print(sess.run(acc, feed_dict={X:mnist.test.images, y_true:mnist.test.labels}))
```

While this may seem pretty good, we can actually do much better, the best models can get above 99% accuracy.

#### Rename the columns:

```python
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'target']
```

### Step 2: Split:

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

### Step 3: Estimator

```python
Image(filename='Image/inputs_to_model_bridge.jpg')#inputs_to_model_bridge.jpg
```

#### 3.1: Feature columns:

- __tf.estimator__a high-level TensorFlow API that greatly simplifies machine learning programming.

__Estimators encapsulate the following actions:__

- training
- evaluation
- prediction
- export for serving  

__Estimators provide a safe distributed training loop that controls how and when to:__  
Estimators build the graph for you.  
build the graph  
initialize variables  
load data  
handle exceptions  
create checkpoint files and recover from failures  
save summaries for TensorBoard 

#### 3.2: Steps:

- A TensorFlow program relying on a pre-made Estimator typically consists of the following four steps:
    - 1. Write one or more dataset importing functions.
    - 2. Define the feature columns. 
    - 3. Instantiate the relevant pre-made Estimator.
    - 4. Call a training, evaluation, or inference method.

```python
Image(filename='Image/tf.png')
```

- Step 1: Feature Columns:

```python
import tensorflow as tf
from platform import python_version
```

```python
print(python_version())
```

```python
print(tf.__version__)
```

```python
X.columns
```

```python
feat_cols = []

for col in X.columns:
    feat_cols.append(tf.feature_column.numeric_column(col))
```

```python
feat_cols
```

- Step 2: Input Function:

```python
input_func = tf.estimator.inputs.pandas_input_fn(x=X_train, y=y_train, batch_size=10, num_epochs=5, shuffle=True)
```

```python
# DNN: Deep neuron network.
classifier = tf.estimator.DNNClassifier(hidden_units=[10, 20, 10], n_classes=3, feature_columns=feat_cols)
```

```python
classifier.train(input_fn=input_func, steps=50)
```

- Step 3: Predict on test case:

Based on every item in test case.

```python
pred_fn = tf.estimator.inputs.pandas_input_fn(x=X_test, y=y_test, batch_size=len(X_test), shuffle=False)
```

```python
classifier.predict(input_fn=pred_fn)
```

```python
note_predictions = list(classifier.predict(input_fn=pred_fn))
```

```python
note_predictions[0]
```

```python
final_preds = []
for pred in note_predictions:
    final_preds.append(pred['class_ids'][0])
```

- Step 4: Evaluation:

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, final_preds))
```

```python
Image(filename='Image/cm.png')
# CM:
```

```python
print(classification_report(y_test, final_preds))
```
